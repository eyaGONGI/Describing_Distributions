# Describing Distributions kurtic, Platykurtic)

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('C:/Users/Dell/Downloads/medical_data.csv')
df.head(2)
```

## VitD_levels

```python
df.columns
```

```python
mean = df['VitD_levels'].mean()
median = df['VitD_levels'].median()
std = df['VitD_levels'].std()
mean_plus_one = mean +1
mean_minus_one = mean - 1 
```

```python
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot( df['VitD_levels'],ax=ax,stat='probability',kde=True)
#Add a line for the mean (red)
ax.axvline(mean, ls='--', color='red', label= f"Mean = {mean:.2f}") 
#Add a line for the median (green)
ax.axvline(median, ls='--', color='green', label= f"Median = {median:.2f}") 
#Add a line for for +1 std from the mean (black)
ax.axvline(mean_plus_one, color = 'black',label=f'+1 std = {mean_plus_one:,.2f}')
#Add a line for the - 1 std from the mean (black)
ax.axvline(mean_minus_one, color = 'black', label = f'-1 std = {mean_minus_one:,.2f}')
#Highlight the range between +1 and =1 std (yellow)
ax.axvspan(mean_plus_one, mean_minus_one, color = 'yellow',zorder=0)

ax.set_title('VitD Levels')
ax.legend();
plt.show()

### Calculate Kurtsis & Skew
kurtsis = stats.kurtosis(df['VitD_levels'], fisher = False)
print(f'Kurtosis = {kurtsis:.2f}')

skew = stats.skew(df['VitD_levels'])
print(f'Skew = {skew:.2f}')
```

* The distribution is  Continuous* The distribution doesn't have a  sew 
* )* It is  (Mesokurtc)
